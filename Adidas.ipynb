```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

sales = [220, 190, 190, 220, 210, 180,
220, 260, 230, 180, 170, 190]

# Plot
plt.figure(figsize=(10,5))
plt.plot(months, sales, marker='o', linewidth=3)

# Labels
plt.title("Trend Line of Unit Sales")
plt.xlabel("Month")
plt.ylabel("Units Sold")

# Show values
for i, value in enumerate(sales):
 plt.text(i, value+3, str(value))

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Online', 'Outlet', 'In-store']
values = [939, 850, 690]

# Plot
plt.figure(figsize=(7,7))
plt.pie(values,
labels=categories,
autopct='%1.1f%%',
startangle=90)

# Title
plt.title("Units Sold vs Sales Category")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
regions = ['West', 'Northeast', 'Southeast', 'South', 'Midwest']
sales = [0.27, 0.19, 0.16, 0.14, 0.14]

# Plot
plt.figure(figsize=(8,5))
plt.barh(regions, sales)

# Labels
plt.title("Region vs Sales")
plt.xlabel("Total Sales (bn)")
plt.ylabel("Region")

# Show values
for i, value in enumerate(sales):
 plt.text(value + 0.005, i, str(value))

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
products = [
    "Men's Street\nFootwear",
    "Women's\nApparel",
    "Men's Athletic\nFootwear",
    "Women's Street\nFootwear",
    "Men's\nApparel",
    "Women's Athletic\nFootwear"
]

sales = [209, 179, 154, 128, 124, 107]

# Plot
plt.figure(figsize=(10,5))
plt.bar(products, sales)

# Labels
plt.title("Products and Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales Amt")

# Values on bars
for i, value in enumerate(sales):
    plt.text(i, value + 3, str(value) + "M", ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
products = [
    "Men's Street Footwear",
    "Women's Apparel",
    "Men's Athletic Footwear",
    "Women's Street Footwear",
    "Men's Apparel",
    "Women's Athletic"
]

profit = [83, 69, 52, 45, 45, 39]

# Plot
plt.figure(figsize=(9,5))
plt.barh(products, profit)

# Labels
plt.title("Product vs Profit")
plt.xlabel("Total Profit")
plt.ylabel("Product")

# Show values
for i, value in enumerate(profit):
    plt.text(value + 1, i, str(value) + "M")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sellers = ['West Gear', 'Foot Locker', 'Sports Direct',
           "Kohl's", 'Walmart', 'Amazon']

units = [625, 604, 558, 287, 206, 198]

# Plot
plt.figure(figsize=(8,5))
plt.barh(sellers, units)

# Reverse order like funnel style
plt.gca().invert_yaxis()

# Labels
plt.title("Units Sold by Seller")
plt.xlabel("Units Sold")

# Values
for i, value in enumerate(units):
    plt.text(value + 5, i, str(value) + "K")

plt.show()
```
